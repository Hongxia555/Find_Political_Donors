```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_table('/Users/Hongxia/Desktop/indiv18/itcont.txt',header=None,sep='|',usecols=[0,10,13,14,15])
```

```python
data.columns = ["CMTE_ID","ZIP_CODE","TRANSACTION_DT","TRANSACTION_AMT","OTHER_ID"]
data=data.head(10)
```

```python
d1 = {}
result1 = pd.DataFrame(columns=['CMTE_ID', 'ZIP_CODE','MEDIAN','COUNT','SUM'])   # CMTE_ID|ZIP_CODE|MEDIAN|#OFRECORDS(length of list)|SUMOFAMOUNT(sum of list)
for i in range(len(data)): # read in one line
    
    # check if it is valid row or not
    # check the length and the format of CMTE_ID
    # check the lenght of ZIP_CODE
    # check if TRANSACTION_AMT is empty
    # check if OTHER_ID is not empty
    if len(data.iloc[i,][0])!=9 or data.iloc[i,][0][:3]!='C00' or len(data.iloc[i,][1])<5 or (data.iloc[i,][3] is None) or not str(data.iloc[i,][4])=='nan':
        continue 
    else:
        # get name,zipcode,date, amount from this row
        name = data.iloc[i,][0]
        zipcode = data.iloc[i,][1][:5]
        amount = data.iloc[i,][3]
        other = data.iloc[i,][4] 

        # put new line into the data structure
        if name not in d1.keys():
            d1[name] = {}
            d1[name][zipcode] = [amount]
            result1.loc[i] = [name, zipcode, np.median(d1[name][zipcode]), 
                              len(d1[name][zipcode]),sum(d1[name][zipcode])]
        else:
            if zipcode not in d1[name].keys():
                d1[name][zipcode] = [amount]
                result1.loc[i] = [name, zipcode, np.median(d1[name][zipcode]), 
                              len(d1[name][zipcode]),sum(d1[name][zipcode])]
            else:
                d1[name][zipcode].append(amount)
                result1.loc[i] = [name, zipcode, np.median(d1[name][zipcode]), 
                              len(d1[name][zipcode]),sum(d1[name][zipcode])]
```

```python
# save as txt file
np.savetxt(r'/Users/Hongxia/Desktop/medianvals_by_zip.txt', result1.values, fmt='%s|%s|%d|%d|%d',
           delimiter='|', header='CMTE_ID|ZIP_CODE|MEDIAN|COUNT|SUM')
```
